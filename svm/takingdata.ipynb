```python
import csv

# Specify the CSV file path
csv_file_path = 'parkinsons.csv'

# Specify the desired order of columns
column_order = [
    'MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)', 'MDVP:Jitter(Abs)', 'MDVP:RAP',
    'MDVP:PPQ', 'Jitter:DDP', 'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'MDVP:APQ', 'Shimmer:DDA',
    'Shimmer:APQ3', 'Shimmer:APQ5', 'NHR', 'HNR', 'RPDE', 'DFA', 'D2', 'PPE'
]

# Read the CSV file
with open(csv_file_path, 'r') as file:
    reader = csv.DictReader(file)
    
    # Print the column values in the specified order
    
    for row in reader:
        data =[]
        for column_name in column_order:
            value = row[column_name]
            data.append(value)
        print(data)   
```
